```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
depth = [110, 290, 470, 650, 830, 1001]

cifar10_sgd_error = [100 - 94.75, 100 - 94.95, 100 - 95.08, 100 - 95.13, 100 - 95.00, 100 - 95.16]
cifar10_srsgd_lin_error = [100 - 95.07, 100 - 95.63, 100 - 95.82, 100 - 96.00, 100 - 95.91, 100 - 96.13]
cifar10_srsgd_exp_error = [100 - 95.00, 100 - 95.50, 100 - 95.51, 100 - 95.60, 100 - 95.71, 100 - 95.87]

cifar10_sgd_bar = [0.14, 0.23, 0.10, 0.14, 0.23, 0.19]
cifar10_srsgd_lin_bar = [0.13, 0.15, 0.09, 0.07, 0.10, 0.07]
cifar10_srsgd_exp_bar = [0.47, 0.18, 0.19, 0.13, 0.13, 0.10]

cifar10_sgd_upper = []
cifar10_sgd_lower = []

cifar10_srsgd_lin_upper = []
cifar10_srsgd_lin_lower = []

cifar10_srsgd_exp_upper = []
cifar10_srsgd_exp_lower = []

for i in range(len(cifar10_sgd_error)):
    cifar10_sgd_upper.append(cifar10_sgd_error[i] + cifar10_sgd_bar[i])
    cifar10_sgd_lower.append(cifar10_sgd_error[i] - cifar10_sgd_bar[i])
    
    cifar10_srsgd_lin_upper.append(cifar10_srsgd_lin_error[i] + cifar10_srsgd_lin_bar[i])
    cifar10_srsgd_lin_lower.append(cifar10_srsgd_lin_error[i] - cifar10_srsgd_lin_bar[i])
    
    cifar10_srsgd_exp_upper.append(cifar10_srsgd_exp_error[i] + cifar10_srsgd_exp_bar[i])
    cifar10_srsgd_exp_lower.append(cifar10_srsgd_exp_error[i] - cifar10_srsgd_exp_bar[i])
```

```python
fig, ax = plt.subplots()
# ax.plot(m, acc_x_vx, label='use x, v0=x')
# ax.plot(m, acc_v_vx, label='use v, v0=x')
ax.plot(depth, cifar10_sgd_error, 'r-o', label='SGD')
plt.fill_between(depth, cifar10_sgd_lower, cifar10_sgd_upper,
                 color='red', alpha=0.2)

ax.plot(depth, cifar10_srsgd_lin_error, 'b-o', label='SRSGD-linear')
plt.fill_between(depth, cifar10_srsgd_lin_lower, cifar10_srsgd_lin_upper,
                 color='blue', alpha=0.2)

ax.plot(depth, cifar10_srsgd_exp_error, 'g-o', label='SRSGD-exponential')
plt.fill_between(depth, cifar10_srsgd_exp_lower, cifar10_srsgd_exp_upper,
                 color='green', alpha=0.2)

ax.legend(loc='upper left')
xsticks = [110, 290, 470, 650, 830, 1001]
plt.setp(ax, xticks=depth, xticklabels=depth)
ax.set_ylim(3.7)
ax.grid()
ax.legend()
fig.show()
fig.savefig('./figures/cifar10_depth_vs_error_srsgd.pdf')
```
